To mine the data from this simulations

```python
import json
import numpy as np
from matplotlib import pyplot as plt
import plotly.graph_objects as go
from bisect import bisect_left

import sys, os
sys.path.append(".")
current_path = os.getcwd()
```

```python

# --------------------------------------------------------------------------- setting the directory
os.chdir(current_path)
os.chdir("../res")

# --------------------------------------------------------------------------- selecting the data files
files     = os.listdir()
files_sel = [] 
for i in files:
    if i.find('json') > 0:
        files_sel.append(i)

# --------------------------------------------------------------------------- importing data 
simulation_results = {}

for i in files_sel:
    with open(i, 'r', encoding="utf-8") as jsonfile:
        data = json.load(jsonfile)
        simulation_results[data['parameters']['ID']] = data
```

```python
# plotting
ID = 1
fig = go.Figure()
mover_stop_idx = (simulation_results[ID]['parameters']['stop_mover_index'])
x     = simulation_results[ID]['data']['t']
wx_og = simulation_results[ID]['data']['wx_og']
wx    = simulation_results[ID]['data']['wx']  
start_idx = 9

# get part where mover stops
fps = (simulation_results[ID]['parameters']['fps'])
dt = 1/fps
offset_time = simulation_results[ID]['parameters']['offset_time']
stop_mover_index = simulation_results[ID]['parameters']['stop_mover_index']
stop_mover_time = offset_time + dt*stop_mover_index
time_index_stop = bisect_left(x, stop_mover_time)


fig.add_trace(go.Scatter(
        x = x[start_idx:],
        y = wx_og[start_idx:],
        name = str("Experiment")
        ))

fig.add_trace(go.Scatter(
            x = x[start_idx:],
            y = wx[start_idx:],
            name = 'Simulation'
            ))

fig.add_trace(go.Scatter(
            x = [x[time_index_stop],x[time_index_stop]],
            y = [0, 1.1*max(wx[start_idx:])],
            name = 'When the target moves on its own'
            ))
    
fig.update_xaxes(
        title_text = "Time [s]",
        title_font = {"size": 20},
        title_standoff = 25
        )

fig.update_yaxes(
        title_text = "Velocity [rad/s]",
        title_font = {"size": 20},
        title_standoff = 25)
fig.show()

# plt.legend()

    
```
